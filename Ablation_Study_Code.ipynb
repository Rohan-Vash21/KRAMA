PARTICULAR LANGUAGE

```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

import pandas as pd

file_path = "TIER_4_FINAL_RESULTS_PATH_OF_A_LANGUAGE_FOR_A_PARTICULAR_MODEL"
df = pd.read_csv(file_path)

columns_to_clean = ['Pred_Logical', 'Pred_Balanced', 'Pred_Creative', 'True_Label']
for col in columns_to_clean:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip().str.upper()

total_records = len(df)

if total_records == 0:
    print("The dataset is empty.")
else:
    prediction_columns = ['Pred_Logical', 'Pred_Balanced', 'Pred_Creative']

    print(f"Total Records: {total_records}\n")

    for pred_col in prediction_columns:
        if pred_col not in df.columns:
            print(f"Column '{pred_col}' not found in the dataset.\n")
            continue

        correct_count = (df[pred_col] == df['True_Label']).sum()
        incorrect_count = total_records - correct_count

        correct_percent = (correct_count / total_records) * 100
        hallucination_percent = (incorrect_count / total_records) * 100

        print(f"--- Results for {pred_col} ---")
        print(f"Correct (%):       {correct_percent:.2f}% ({correct_count}/{total_records})")
        print(f"Hallucination (%): {hallucination_percent:.2f}% ({incorrect_count}/{total_records})\n")
```
